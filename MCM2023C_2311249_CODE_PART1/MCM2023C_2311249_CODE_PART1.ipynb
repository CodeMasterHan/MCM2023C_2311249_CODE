```python
l=[5988, 1627, 2027, 2453, 6662, 1115,1643,  1759, 3758,291, 1505, 3371, 1976, 2952,4438, 2019, 112, 4157, 6663,  3295, 2511, 694,  1039, 288, 2074, 434]
larray = np.array(l)
lnx = np.log(np.array(l))
standard_l = (larray-112)/(6663-112)
def wordVar(word):
    wordArray = []
    for letter in word:
        try:
            wordArray.append(standard_l[ord(letter)-97])
        except IndexError:
            if(letter==' '):
                continue
    nwordArray = np.array(wordArray)
    return np.var(nwordArray)
def frequencyGap(word):
    wordArray = []
    for letter in word:
        try:
            wordArray.append(standard_l[ord(letter)-97])
        except IndexError:
            if(letter==' '):
                continue
    nwordArray = np.array(wordArray)
    return np.max(nwordArray)-np.min(nwordArray)
def frequencyGap2(word):
    wordArray = []
    for letter in word:
        try:
            wordArray.append(standard_l[ord(letter)-97])
        except IndexError:
            if(letter==' '):
                continue
    nwordArray = np.array(wordArray)
    max1= np.max(nwordArray)
    min1= np.min(nwordArray)
    wordArray.remove(max1)
    wordArray.remove(min1)
    nwordArray = np.array(wordArray)
    max2= np.max(nwordArray)
    min2= np.min(nwordArray)
    return max1+max2-min1-min2
def frequencySum(word):
    wordArray = []
    for letter in word:
        try:
            wordArray.append(standard_l[ord(letter)-97])
        except IndexError:
            if(letter==' '):
                continue
    nwordArray = np.array(wordArray)
    return np.sum(nwordArray)
```
